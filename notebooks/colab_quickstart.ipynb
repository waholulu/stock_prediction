## Results

```python
import pandas as pd

binary = pd.read_csv("results/binary_metrics.csv")
print("Binary classification — fold-by-fold metrics")
print(binary.to_string(index=False))
print(f"\nMean: acc={binary['accuracy'].mean():.4f}  MCC={binary['mcc'].mean():.4f}  AUC={binary['auc'].mean():.4f}")
```

```python
import matplotlib.pyplot as plt

equity = pd.read_csv("results/equity_curve.csv", parse_dates=["date"])

fig, ax = plt.subplots(figsize=(12, 5))
ax.plot(equity["date"], equity["equity"], label="Strategy (net of costs)")
ax.plot(equity["date"], equity["bh_equity"], label="Buy & Hold", linestyle="--")
ax.set_xlabel("Date")
ax.set_ylabel("Equity (starting at 1.0)")
ax.set_title("OOS Equity Curve vs Buy-and-Hold")
ax.legend()
fig.tight_layout()
plt.show()
```
